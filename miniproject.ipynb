```python
import pandas as pd
from datetime import datetime as dt
pd.options.display.max_colwidth = 200
```

```python
df1 = pd.read_csv("fake_data_address.csv",sep=',')
df2 = pd.read_csv("fake_data_dob.csv",sep=',')
```

```python
df1.head(-5)
```

```python
df1['full_name'] = df1['last_name'] + ', ' + df1['first_name'] + ' ' + df1['middle_name']
df1.head()
```

```python
df2 = df2.rename(columns={'last_name, first_name middle_name':'full_name'})
df2.head(-5)
```

```python
#df3 = pd.merge(df1,df2,on='full_name')
#df3
```

```python
df1['address_line_1'] = df1['Line 1: address Line 2: address2'].str.extract(r'^Line 1: (.*)(?:Line 2: )')
df1['address_line_2'] = df1['Line 1: address Line 2: address2'].str.extract(r'Line 2: (.*$)')
df1
```

```python
df3 = pd.merge(df1,df2,on='full_name')
df3 = df3.drop(['Line 1: address Line 2: address2','full_name'],axis=1)
df3
```

```python
df3['state'] = df3['state-city'].str.extract(r'(^.*)-')
df3['city'] = df3['state-city'].str.extract(r'-(.*$)')
df3 = df3.drop('state-city',axis=1)
df3
```

```python
df3['zip'] = df3['zip'].astype(int)
df3
```

```python
df3["dob"] = pd.to_datetime(df3["dob"])
df3
```

```python
now = dt.now()
```

```python
df3['age'] = (now - df3['dob']).astype('<m8[Y]')
df3.head()
```

```python
age_df = df3.loc[(df3["age"] >= 40) &
                (df3["age"] <= 60) , :]
age_df
```

```python
for index, row in age_df.iterrows():
    print(f"-------------{index}-------------")
    print(f"{row['first_name']} {row['middle_name']} {row ['last_name']}")
    print(f"{row['address_line_1']}")
    if row['address_line_2'] != "":
        print(f"{row['address_line_2']}")
    print(f"{row['city']}, {row['state']} {row['zip']}")
```
